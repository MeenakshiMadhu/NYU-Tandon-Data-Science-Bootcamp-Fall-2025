### 1. Weekday pedestrian counts
#### Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.

```python
# Filter all weekdays
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekday_data = df[df['DayOfWeek'].isin(weekdays)]

# Group by weekday and sum pedestrian counts
weekday_counts = weekday_data.groupby('DayOfWeek')['Pedestrians'].sum().reindex(weekdays)
weekday_counts
```

```python
# Plot
plt.figure(figsize=(10,5))
plt.plot(weekday_counts.index, weekday_counts.values, marker='o', linewidth=2)
plt.title('Pedestrian Counts by Weekday')
plt.xlabel('Day of Week')
plt.ylabel('Total Pedestrian Count')
plt.grid(True, alpha=0.6)
plt.show()
```

#### Observations:
There's a clear dip in the number of pedestrians or visitors in the middle of the week and a sharp incline at the end. This could probably be due to tourists planning their trips more around the weekend, and less during the middle of the work week.<br>
Fridays have the highest count of pedestrians, probably due to the beginning of the weekend. <br>


### 2. 2019 Weather influence on pedestrians
Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

```python
# Filter for Brooklyn Bridge and 2019
brooklyn_2019 = df[(df['location'] == 'Brooklyn Bridge') & (df['Year'] == 2019)].copy()
# Check available weather summaries
brooklyn_2019['weather_summary'].value_counts().head()
```

```python
# Average pedestrian count by weather
weather_counts = (
    brooklyn_2019.groupby('weather_summary')['Pedestrians']
    .mean()
    .sort_values(ascending=False)
)
```

```python
# Plot average pedestrian counts by weather
plt.figure(figsize=(12,6))
sns.barplot(x=weather_counts.index, y=weather_counts.values, hue=weather_counts.index, dodge=False, legend=False)
plt.title('Average Pedestrian Counts by Weather (Brooklyn Bridge, 2019)')
plt.xlabel('Weather Summary')
plt.ylabel('Average Pedestrian Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Observations:
The highest traffic was during days with partly cloudy or clear skies. This is enjoyable weather and hence preferred by most people to be outside. Extreme conditions like rain, snow, or even sleet show less traffic. <br>
The least traffic was during the "night" weather summary values, even when it had pleasant weather conditions. This is expected since more visitors come during the day.

```python
# Correlation matrix between weather, temp, precip, pedestrians

## Arranging the weather conditions roughly from mild to extreme
custom_weather_map = {
    'clear-day': 0,
    'clear-night': 1,
    'partly-cloudy-day': 2,
    'partly-cloudy-night': 3,
    'cloudy': 4,
    'wind': 5,    
    'rain': 6,
    'fog': 7,
    'snow': 8,
    'sleet': 9
}
brooklyn_2019['Weather_Code'] = brooklyn_2019['weather_summary'].map(custom_weather_map)
```

```python
corr_matrix = brooklyn_2019[['Pedestrians', 'temperature', 'precipitation', 'Weather_Code']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix (Brooklyn Bridge, 2019)')
plt.show()

corr_matrix
```

#### Observations:
There is positive correlation between traffic and temperature, indicating that people prefer being outside at higher temperatures. <br>
There is observable negative correlation between precipitation and pedestrian counts. This is an expected outcome, since there will be fewer pedestrians traffic when it is raining or snowing. <br>
There is also a major negative correlation between the weather code and pedestrian traffic. This indicates that the higher the weather code, the less the traffic. Higher weather codes were given for more extreme conditions, which makes this a reasonable observation.

### 3. Pedestrian Activity throughout the day
Implement a custom function to categorize time of day into morning, afternoon, evening, and night, and create a new column in the DataFrame to store these categories. Use this new column to analyze pedestrian activity patterns throughout the day.

```python
# Define custom categorization function
def categorize_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
```

```python
# Create new column
df['TimeOfDay'] = df['Hour'].apply(categorize_time_of_day)

# Analyze average pedestrian count by time of day
tod_counts = df.groupby('TimeOfDay')['Pedestrians'].mean().reindex(['Morning', 'Afternoon', 'Evening', 'Night'])
tod_counts
```

```python
# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=tod_counts.index, y=tod_counts.values, color='teal')
plt.title('Average Pedestrian Activity by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Average Pedestrian Count')
plt.show()
```

### Observations:
The highest traffic is observed during the afternoon, followed by the evening, and the least during nighttime.
